```python
bins = 50
binRange = [0,4000]

binedges = np.linspace(binRange[0],binRange[1], bins)

CsInt1, CsBins1 = np.histogram(Ch0Int,binedges)
CsInt2, CsBins2 = np.histogram(Ch1Int,binedges)

bckInt1, bckBins1 = np.histogram(bckCh0Int,binedges)
bckInt2, bckBins2 = np.histogram(bckCh1Int,binedges)



fig, ax = plt.subplots(2,2, figsize = (10,10))

ax[0,0].step(CsBins1[:-1], CsInt1, label = 'Cs-137 not normalized')
ax[0,0].step(bckBins1[:-1],bckInt1, label = 'Background not normalized')

ax[0,1].step(CsBins2[:-1], CsInt2, label = 'Cs-137 not normalized')
ax[0,1].step(bckBins2[:-1],bckInt2, label = 'Background not normalized')



ax[1,0].step(CsBins1[:-1], CsInt1/runTime, label = 'Cs-137 normalized')
ax[1,0].step(bckBins1[:-1],bckInt1/bckRunTime, label = 'Background normalized')

ax[1,1].step(CsBins2[:-1], CsInt2/runTime, label = 'Cs-137 not normalized')
ax[1,1].step(bckBins2[:-1],bckInt2/bckRunTime, label = 'Background not normalized')

ax[0,0].legend(loc = 'best')
ax[0,1].legend(loc = 'best')
ax[1,1].legend(loc = 'best')
ax[1,0].legend(loc = 'best')


plt.show()

fig, ax = plt.subplots(1,2)
plt.tight_layout()

normalize = False
LogScale = True

ax[0].hist(CsBins1[:-1],bins =CsBins1,density = normalize,weights=CsInt1/runTime,histtype = 'step',label = 'Cs-137 Ch 0', log = LogScale)
ax[1].hist(CsBins2[:-1],bins =CsBins2,density = normalize,weights=CsInt2/runTime,histtype = 'step',label = 'Cs-137 Ch 1', log = LogScale)

ax[0].hist(bckBins1[:-1],bins =bckBins1,density = normalize,weights=bckInt1/bckRunTime,histtype = 'step',label = 'Background Ch 0', log = LogScale)
ax[1].hist(bckBins2[:-1],bins =bckBins2,density = normalize,weights=bckInt2/bckRunTime,histtype = 'step',label = 'Background Ch 1', log = LogScale)


# ax[0].hist(Ch0Int,bins = (binedges), log = True, histtype = 'step', label = 'Cs-137')
# ax[0].hist(bckCh0Int,bins = (binedges), log = True, histtype = 'step', label = 'Background')

# ax[1].hist(Ch1Int,bins = (binedges), log = True, histtype = 'step', label = 'Cs-137')
# ax[1].hist(bckCh1Int,bins = (binedges), log = True, histtype = 'step', label = 'Background')

ax[0].set_xlabel(r'Ch 0 Integral (ADU$\mu$s)')
ax[1].set_xlabel(r'Ch 1 Integral (ADU$\mu$s)')

ax[0].set_ylabel('Counts/bin/s')
ax[1].set_ylabel('Counts/bin/s')

ax[0].legend(loc = 'best')
ax[1].legend(loc = 'best')
plt.savefig('/home/nick/PhD/KDK+/Large_LSC_testing/Initial_testing_old_DAQ/2025_01_16/figures/1d_hist_logscale.png', dpi = 400)

plt.show()
```
